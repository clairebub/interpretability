```python
from PIL import Image

im = Image.open('../results/gram_cam/0-ISIC_0024707-resnet34-gradcam-module.layer4-BKL.png') # Can be many different formats.
pix = im.load()
img_size = im.size

for x in range(img_size[0]):
    for y in range(img_size[1]):
    
        if pix[x,y][0] < 150:
            pix[x,y] = (0,0,0)
        else:
            pix[x,y] = (255,255,255)

imgplot = plt.imshow(im)
plt.show()
```
